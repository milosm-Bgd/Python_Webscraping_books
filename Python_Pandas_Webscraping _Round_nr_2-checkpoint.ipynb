```python
class Vehicle:
    def __init__(self, brand, model, type_of_the_car):
        #attributes
        self.brand = brand
        self.model = model
        self.type_of_the_car = type_of_the_car
        self.gas_tank_size = 14
        self.fuel_level = 5

    #definition
    def fuel_up(self):
        self.fuel_level = self.gas_tank_size
        print('Gas tank is now full!')
    def drive(self):
        print(f'The {self.model} is now driving')
    def update_fuel_level(self, some_level):
        if some_level <= self.gas_tank_size:
            self.fuel_level = some_level
        else:
            print('Exceeded capacity')
    def __str__(self):
        return str('I\'m a a vehicle')
```

```python
#Define the Book class
class Book():
    def __init__(self, title, price, read):
	    self.title = title
	    self.price = price
	    self.read = read 

    def __repr__(self):
	    return f'Book({self.title}, {self.price}£, Have I read the book: {self.read})'
```

```python
base_url = 'https://books.toscrape.com/catalogue/page-{}.html'
has_next = True
n = 1
all_books= []

while (has_next):   #we loop through the PAGES
  
  scrape_url = base_url.format(n)
  res = requests.get(scrape_url)

  soup = BeautifulSoup(res.text, "lxml")

  books = soup.select(".product_pod")
  print(f'There are len{books} books on the page {n}.')

  for book in books:

    title = book.select('a')[1]['title']
    price = float(book.select('.price_color')[0].getText().split('£')[1])
    #or call the function we wrote
    #price = get_price(book.select('.price_color')[0].getText(), '£')

    #lEts print what we have
    print(title, price)

    #creation an Object of a Class Book
    book_object = Book(title, price, random.choice([True, False]))

    #Q:to which class / data structure does the 'append' method belong to?
    all_books.append(book_object)

  #do we have additional pages
  has_next_button_list = soup.select('.next')

  if not has_next_button_list:
    has_next = False
    print('Loop is over')
  else:
    print('Continue to next page...')
     
  n+=1
```

```python
def sum_of_books_price(list_of_books):
  sum_of_all_books = 0
  sum_of_books_i_have_read = 0
  for book in list_of_books:
    sum_of_all_books += book.price
    if book.read:
      sum_of_books_i_have_read += book.price

  return sum_of_all_books , sum_of_books_i_have_read
```

### Series

The first main data type we will learn about for pandas is the Series data type. Let's import Pandas and explore the Series object.

A Series is very similar to a NumPy array (in fact it is built on top of the NumPy array object). What differentiates the NumPy array from a Series, is that a Series can have axis labels, meaning it can be indexed by a label, instead of just a number location. It also doesn't need to hold numeric data, it can hold any arbitrary Python Object.

Let's explore this concept through some examples:

```python
import numpy as np
import pandas as pd
```

We can convert a list, numpy array or dictionary to a Series:

```python
labels = ['a', 'b', 'c']
my_list = [10, 20, 30]
arr = np.array([10, 20, 30]) # arrays -> they know how to perform: element wise operation
d = {'a':10, 'b':20, 'c':30}
```

##### Merging, Joining and Concatenating
There are 3 main ways of combining DataFrames together: Merging, Joining and Concatenating. Here we will discuss these 3 methods with examples.

```python
#create a funcion to calculate missing values for each column in the df
def table_view_of_missing_values(df):
    total = df.isnull().sum().sort_values(ascending=False)
    percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
    missing_data = pd.concat([total,percent], axis=1, keys=['Total','Percent'])
    return missing_data #[missing_data['Total'] > 0]
```

###### Business Question:

OPTIONAL

We are interested in the Job Positions: CHIEF, CAPTAIN and LIEUTENANT.

For these Job Positions, please calculate the min, max and mean BasePay, as well as the max OtherPay, throughout the years.

How did the salaries develop over the years? (Can we explain why?)

```python
job_titles_dictionary = sal['JobTitle'].value_counts().to_dict()

mapping_dictionary = {}

for key in job_titles_dictionary.keys():
    # print(key)
    if 'chief' in key.lower():
        mapping_dictionary[key] = 'CHIEF'
    elif 'captain' in key.lower():
        mapping_dictionary[key] = 'CAPTAIN'
    elif 'lieutenant' in key.lower():
        mapping_dictionary[key] = 'LIEUTENANT'
    else:
        mapping_dictionary[key] = 'Irrelevant_for_research'
    
```

```python
sal.head()
```

```python
sal['JobTitleMainGroup'] = sal['JobTitle'].map(mapping_dictionary)
```

```python
sal.head()
```

```python
pd.options.display.float_format = '${:,.2f}'.format
sal.groupby(['JobTitleMainGroup','Year']).agg({'BasePay': ['mean','min','max'], 'OtherPay': ['max']}).drop('Irrelevant_for_research', axis=0)
```

```python
plot_df = sal.groupby(['JobTitleMainGroup', 'Year']).agg({'BasePay': ['mean', 'min', 'max'], 'OtherPay': ['max'] }).drop('Irrelevant_for_research', axis=0)['BasePay']
```

```python
import matplotlib.pyplot as plt
# documenation https://matplotlib.org/
plt.rcParams["figure.figsize"]=(12, 8)
import seaborn as sns
sns.set_theme()
plot_df.plot(marker="s",ms=10, xticks=[2011, 2012, 2013, 2014]);
```
